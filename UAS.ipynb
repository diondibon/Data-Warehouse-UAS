## Import Pandas

```python
import matplotlib
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

## Memuat File CSV ke dalam DataFrame

```python
dataset = pd.read_csv('dataset.csv')
dataset
```

#### menampilkan Total Buku Masuk dan Total Buku Keluar dari dataset untuk tiap Tahun
Dataset akan dikelompokkan terlebih dahulu berdasarkan Tahun, selanjutnya akan dicari berapa Total Buku Masuk dan Total Buku Keluar

```python
dataset.groupby('Tahun').apply(lambda df: df[ df['Total Buku Masuk'] == df['Total Buku Masuk'].max() ])[['Total Buku Keluar']]
```

#### Menampilkan data dari Perpustakaan untuk tiap Tahun di Bulan Tertentu dengan Total Buku Keluar tertinggi dengan Lokasinya

```python
dataset.groupby(['Tahun','Bulan']).apply(lambda df: df[ df['Total Buku Keluar'] == df['Total Buku Keluar'].max() ])[['Lokasi']]
```

### Menampilkan Min dan Max Total Buku Keluar Di Seluruh Lokasi Perpustakaan

```python
dataset.groupby('Lokasi')['Total Buku Keluar'].agg([min, max])
```

### Menampilkan Min dan Max Total Buku Masuk Di Seluruh Lokasi Perpustakaan

```python
dataset.groupby('Lokasi')['Total Buku Masuk'].agg(['min', 'max'])
```

#### menampilkan rata-rata Total Buku Keluar di setiap Lokasi secara ascending

```python
dataset.groupby('Lokasi')['Total Buku Keluar'].mean().sort_values(ascending=False)
```
